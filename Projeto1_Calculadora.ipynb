```python
# Minha primeira calculadora em Python :)

while True:
    print("\n=== Calculadora ===")

    # aqui o usuário digita os números (tem que ser número, senão dá erro mesmo)
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))

    # menu de operações
    print("\nEscolha a operação:")
    print("1 - Soma")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")

    opcao = input("Digite o número da operação desejada: ")

    # verificando qual opção foi escolhida
    if opcao == "1":
        resultado = num1 + num2
        print(f"Resultado: {num1} + {num2} = {resultado}")
    elif opcao == "2":
        resultado = num1 - num2
        print(f"Resultado: {num1} - {num2} = {resultado}")
    elif opcao == "3":
        resultado = num1 * num2
        print(f"Resultado: {num1} * {num2} = {resultado}")
    elif opcao == "4":
        if num2 != 0:  # evitar divisão por zero
            resultado = num1 / num2
            print(f"Resultado: {num1} / {num2} = {resultado}")
        else:
            print("Erro: divisão por zero não é permitida.")
    else:
        print("Opção inválida, tente novamente.")

    # perguntar se continua
    continuar = input("\nDeseja realizar outra operação? (s/n): ")
    if continuar.lower() != "s":
        print("Encerrando a calculadora. Até logo!")
        break



```
